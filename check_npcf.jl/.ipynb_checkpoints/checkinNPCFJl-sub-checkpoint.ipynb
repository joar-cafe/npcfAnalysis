## MAPEO DE TREECORR A r1, r2, mu

```python
# coords from tc
nmubins=50
newzeta=np.zeros((nrbins,nrbins,nmubins))
deltar=(rmax-rmin)/nrbins

coords=np.zeros((nrbins,nubins,2*nubins,3))
for i in range(len(corr.logr1d)):
    for j in range(len(corr.u1d)):
        for k in range(len(corr.v1d)):
            coords[i,j,k,:]=np.exp(corr.logr1d[i]),corr.u1d[j],corr.v1d[k]
```

```python
# MAPEO
for i in range(len(corr.logr1d)):
    for j in range(len(corr.u1d)):
        for k in range(len(corr.v1d)):
            #r1=coords[i,j,k,0]
            #r2=coords[i,j,k,0]*coords[i,j,k,1]
            if coords[i,j,k,2]<0:
                mu=0.5*(coords[i,j,k,1]+2*coords[i,j,k,2]-coords[i,j,k,1]*(coords[i,j,k,1]**2))
            else:
                mu=0.5*(coords[i,j,k,1]-2*coords[i,j,k,2]-coords[i,j,k,1]*(coords[i,j,k,1]**2))
            
            #r2index=i+np.rint(np.log(coords[i,j,k,1])/deltar) # log(r2)=log(r1*u)=log(r1)+log(u) which in bins is i+int(log(u)/Delta_r)
            r2index=i*coords[i,j,k,1]
            if r2index<0: r2index=0
            
            muindex=int(np.arccos(mu)*nmubins/2/np.pi)
            #muindex=int((mu+1)*mubins/2) #factor of 2 due to 10 bins from negative values and 10 from possitive values.
            
            newzeta[i,int(r2index),muindex]+=corr.zeta[i,j,k] #r1 bin is the same as before = i and log(r2) can 
```

```python
#simetrize
newzetasym=np.copy(newzeta)

for i in range(len(corr.logr1d)):
    for j in range(len(corr.logr1d)):

        newzetasym[i,j,:]=(newzeta[i,j,:]+newzeta[j,i,:])/2.
```
